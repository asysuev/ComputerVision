```python
import numpy as np
def lm_files_to_dict(dest = 'original'):
    path_to_landmarks = '/Users/asysuev/OneDrive/Study/KULeuven/ComputerVision/Project Data(2)/_Data/Landmarks/'+dest+'/'
    landmark_files = {}
    lm_name = 'landmarks'
    for i in range(1,9):
        key = str(i)
        landmark_files[key]=[]
        for j in range(1,14):
            landmark_files[key].append(path_to_landmarks+lm_name+str(j)+'-'+key+'.txt')
    return landmark_files

def load_landmarks(landmark_files):
    landmarks = {}
    for key in landmark_files:
        landmarks[key]=[]
        for fl in landmark_files[key]:
            with open(fl) as infile:
                data = infile.readlines()
            landmarks_1 = []
            for line in data:
                landmarks_1.append(float(line))
            landmarks[key].append(np.asarray(landmarks_1))
    return landmarks
lm_files = lm_files_to_dict('original')

landmarks = load_landmarks(lm_files)
landmarks
```
